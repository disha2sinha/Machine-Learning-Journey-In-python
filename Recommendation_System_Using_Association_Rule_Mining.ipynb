```
import pandas as pd
from itertools import groupby,combinations
from collections import Counter
df1=pd.read_csv('https://static.turi.com/datasets/millionsong/10000.txt',header=None,sep='\t')
df1.columns=['user_id','song_id','listen_count']
df1.head(10)
```

```
songs=df1.set_index('user_id')['song_id']
songs.head(10)
```

```
print('Unique_users:{0}\tUnique_songs:{1}'.format(len(songs.index.unique()),len(songs.value_counts())))
```

```
song_stats=songs.value_counts().rename('Frequency').to_frame('Frequency')
song_stats['support']=(song_stats['Frequency']/len(set(songs.index)))
song_stats.head(10)
```

```
qualifying_songs=song_stats[(song_stats['support']>=0.001) & (song_stats['Frequency']>=100)].index
song_items=songs[songs.isin(qualifying_songs)]
display('Songs with support >={}:{}   Remaining users:{}'.format(0.001,len(qualifying_songs),len(song_items.index)))
```

```
def get_song_pairs(song_items):
  song_items=song_items.reset_index().as_matrix()
  for song_id,song in groupby(song_items,lambda x:x[0]):
    song_list=[songs[1] for songs in song]
    for song_pair in combinations(song_list,2):
      yield song_pair
```

```
song_stats=song_items.value_counts().rename("Frequency").to_frame("Frequency")
song_stats['Support']=song_stats['Frequency']/len(set(song_items.index))*100
song_pair_gen=get_song_pairs(song_items)
song_pairs=pd.Series(Counter(song_pair_gen)).rename("Frequency").to_frame("FrequencyXY")
song_pairs['SupportXY']=song_pairs['FrequencyXY']/len(qualifying_songs)*100
print("Song Pairs:{}".format(len(song_pairs)))
```

```
song_pairs.head(50)
```

```
song_pairs['SupportXY'].min()
```

```
song_pairs=song_pairs[song_pairs['SupportXY']>=0.1]
print("Song pairs with support>=0.1:{}".format(len(song_pairs)))
```

```
def merge_song_pairs(song_pairs,song_stats):
  return(song_pairs.merge(song_stats.rename(columns={'Frequency':'Frequency_X','Support':'Support_X'}),left_on='Song_X',right_index=True).merge(song_stats.rename(columns={'Frequency':'Frequency_Y','Support':'Support_Y'}),left_on='Song_Y',right_index=True))
```

```
song_pairs=song_pairs.reset_index().rename(columns={'level_0':'Song_X','level_1':'Song_Y'})
song_pairs=merge_song_pairs(song_pairs,song_stats)
song_pairs.head(40)
```

```
song_pairs['ConfidenceXY']=song_pairs['SupportXY']/song_pairs['Support_X']
song_pairs['ConfidenceYX']=song_pairs['SupportXY']/song_pairs['Support_Y']
song_pairs['Lift']=song_pairs['SupportXY']/(song_pairs['Support_X']*song_pairs['Support_Y'])
song_pairs.head(20)
```

```
song_name=pd.read_csv('https://static.turi.com/datasets/millionsong/song_data.csv',index_col='song_id')
song_name.head()
```

```
song_name=song_name.rename(columns={'title':'Song_Title','artist_name':'Singer'})
song_name=song_name.drop(columns={'release','year'})
song_name.head()
```

```
#columns=['Song_X','Song_Y','FrequencyXY','SupportXY','Frequency_X','Support_X','Frequency_Y','Support_Y','ConfidenceXY','ConfidenceYX','Lift']
song_pairs_recommendation=song_pairs.merge(song_name.rename(columns={'Song_Title':'song_X','Singer':'Singer_X'}),left_on='Song_X',right_on="song_id").merge(song_name.rename(columns={'Song_Title':'song_Y','Singer':'Singer_Y'}),left_on='Song_Y',right_on="song_id")
song_pairs_recommendation.head()
```

```
song_pairs_recommendation=song_pairs_recommendation[['song_X','Singer_X','song_Y','Singer_Y','FrequencyXY','Lift']].sort_values('Lift',ascending=False)
song_pairs_recommendation.head(40)
```
