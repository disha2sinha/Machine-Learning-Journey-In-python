## **Loading the Dataset**

```
from google.colab import files
uploaded=files.upload()
```

```
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['primary-tumor.data']))
df.columns=['Class','Age','Sex','Histologic-type','Degree-of-diffe','Bone','Bone-marrow','Lung','Pleura','Peritoneum','Liver','Brain','Skin','Neck','Supraclavicular','Axillar','mediastinum','Abdominal']
df
```

## Dropping rows with null values

```
df.info()
```

```
for column in df[['Sex','Histologic-type','Degree-of-diffe','Skin','Axillar']]:

  df =df[pd.to_numeric(df[column], errors='coerce').notnull()]
  df[column] =df[column].astype('int')
df.info()
```

```
df.reset_index(drop=True)
```

```
df.info()
```

## Finding unique values of columns

```
for column in df.columns:
  print('Column:',column,'\t\t\t\t\tUnique Values:',df[column].unique())
```

## Train Test Split

```
from sklearn.model_selection import train_test_split
X=df.drop(['Class'],axis=1)
y=df['Class']
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

## Hyperparameter Tuning

```
from sklearn.metrics import f1_score
from sklearn import svm
kernels=['linear','poly','rbf','sigmoid']
cs=[0.1,1,10,100,1000]
gammas=[0.1,1,10,100]
for item in kernels:
  if (item=='rbf'):
    for c in cs:
      for g in gammas:
        clf = svm.SVC(kernel=item,C=c,gamma=g)
        clf.fit(X_train, y_train)
        y_pred=clf.predict(X_test)
        accuracy=f1_score(y_test, y_pred,average='weighted') 
        print('Kernel=',item,' C=',c,'gamma=',g,' Accuracy=',accuracy*100,"%")
  elif item=='poly':
    for degree in range(10):
      clf=svm.SVC(kernel=item,degree=degree,gamma=0.1)
      clf.fit(X_train, y_train)
      y_pred=clf.predict(X_test)
      accuracy=f1_score(y_test, y_pred,average='weighted') 
      print('Kernel=',item,'degree=',degree,'gamma=0.1','Accuracy=',accuracy*100,"%")
  else:
    clf=svm.SVC(kernel=item)
    clf.fit(X_train, y_train)
    y_pred=clf.predict(X_test)
    accuracy=f1_score(y_test, y_pred,average='weighted') 
    print('Kernel=',item,'Accuracy=',accuracy*100,"%")
  
```

## Finding Maximum Accuracy Using Jaccard Index

```
clf=svm.SVC(kernel='poly',degree=2,gamma=0.1)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```
from sklearn.metrics import jaccard_similarity_score
jaccard_similarity_score(y_test, y_pred)
```

## F1-Score

```
f1_score(y_test, y_pred, average='weighted')
```

## CLASSIFICATION REPORT

```
from sklearn.metrics import classification_report,confusion_matrix
print (classification_report(y_test, y_pred))
```
